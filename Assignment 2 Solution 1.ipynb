```python
data = []

doc_string = '''
    'data' is basically a list which acts as a hash table by storing the value and it's hash key as tuple.
    
    Insert function :
    - Takes in the argument, compares it's hash value in the list and sorts while insertion.
    - The data is sorted in increasing order.
    
    Search and Delete functions:
    - Search for the data by argument's hash value and returns the result if found. 
    - The delete function further removes the data from the list.    
'''

def insert(ele):
    h = hash(ele)
    if len(data) == 0:
        data.append((h, ele))
        return data
    else:
        for idx, (i, j) in enumerate(data):
            if h < i:
                data.insert(idx, (h, ele))
                return data
            elif h == i:
                return data
            else:
                if idx == len(data) - 1:
                    data.append((h, ele))
                    return data


def search(ele):
    h = hash(ele)
    for idx, (i, j) in enumerate(data):
        if i == h:
            return f"Data Found! {(i, j)}"
        elif i > h:
            return "\nData not found!"


def delete(ele):
    h = hash(ele)
    for i, j in data:
        if i == h:
            data.pop(data.index((i, j)))
            return f"\nThe data {i, j} has been deleted Successfully!"
        elif i > h:
            return f"\nCouldn't delete. Data not found!"


# Driver Code
if __name__ == "__main__":

    i1 = insert("Amit")
    print(i1)
    i2 = insert("Ullas")
    print(i2)
    i3 = insert("Vikas")
    print(i3)

    print(f"\nData List: {data}")

    s1 = search("Ullas")
    print(f"\nSearch Result : {s1}")
    s2 = search("Seema")
    print(s2)

    d1 = delete("Ullas")
    print(d1)
    d2 = delete("Seema")
    print(d2)

    
    ''' The above solution can also be solved using Dictionaries,
        which provide hashable keys, insert, pop and search methods available implicitly.
    ''' 
```
